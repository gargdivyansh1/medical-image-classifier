```python
import sys
import os

sys.path.append(os.path.abspath(os.path.join(os.getcwd(), "..")))

from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
from src.model import build_model
```

```python
train_path = '../data/chest_xray/train'
val_path = '../data/chest_xray/val'
```

```python
datagen = ImageDataGenerator(rescale = 1./255, horizontal_flip = True, zoom_range = 0.2)
```

```python
train_gen = datagen.flow_from_directory(train_path, target_size=(224, 224), class_mode = 'binary', batch_size=32)
```

```python
val_gen = datagen.flow_from_directory(val_path, target_size=(224, 224), class_mode='binary', batch_size=32)
```

```python
model = build_model()
history = model.fit(train_gen, validation_data = val_gen, epochs = 10)
```

```python
model.save('penumonia_model.h5')
```
